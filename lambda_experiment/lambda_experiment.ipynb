```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lambda_00 = pd.read_csv('lambda_00.csv')
lambda_15 = pd.read_csv('lambda_15.csv')
lambda_50 = pd.read_csv('lambda_50.csv')
lambda_85 = pd.read_csv('lambda_85.csv')
lambda_100 = pd.read_csv('lambda_100.csv')

epochs = len(lambda_00)
```

```python
plt.figure(figsize=(6,4))
plt.plot(range(epochs), lambda_100['train'], label='Train')
plt.plot(range(epochs), lambda_100['valid'], label='Valid')
plt.xlabel('Epochs'); plt.ylabel('Lambda')
plt.xlim(0,epochs); plt.ylim(0,50)
plt.legend(loc='upper right')
plt.grid(True, which='both', alpha=0.25)
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,6), sharex=True, sharey=True)

axs[0,0].plot(range(epochs), lambda_00['train'], label=r'Train - $\omega=0.00$')
axs[0,0].plot(range(epochs), lambda_15['train'], label=r'Train - $\omega=0.15$')
axs[0,0].plot(range(epochs), lambda_50['train'], label=r'Train - $\omega=0.50$')
axs[0,0].plot(range(epochs), lambda_85['train'], label=r'Train - $\omega=0.85$')
axs[0,0].plot(range(epochs), lambda_100['train'], label=r'Train - $\omega=1.00$')
axs[0,0].set(ylabel='Loss', title=r'Total Training Loss, $\mathcal{L}_{train}$')
axs[0,0].legend(loc='upper right'); axs[0,0].grid(True, which='both', alpha=0.25)

axs[0,1].plot(range(epochs), lambda_00['valid'], label=r'Valid - $\omega=0.00$')
axs[0,1].plot(range(epochs), lambda_15['valid'], label=r'Valid - $\omega=0.15$')
axs[0,1].plot(range(epochs), lambda_50['valid'], label=r'Valid - $\omega=0.50$')
axs[0,1].plot(range(epochs), lambda_85['valid'], label=r'Valid - $\omega=0.85$')
axs[0,1].plot(range(epochs), lambda_100['valid'], label=r'Valid - $\omega=1.00$')
axs[0,1].set(ylabel='Loss', title=r'Total Validation Loss, $\mathcal{L}_{valid}$')
axs[0,1].legend(loc='upper right'); axs[0,1].grid(True, which='both', alpha=0.25)

axs[1,0].plot(range(epochs), lambda_00['phys'], label=r'Physics - $\omega=0.00$')
axs[1,0].plot(range(epochs), lambda_15['phys'], label=r'Physics - $\omega=0.15$')
axs[1,0].plot(range(epochs), lambda_50['phys'], label=r'Physics - $\omega=0.50$')
axs[1,0].plot(range(epochs), lambda_85['phys'], label=r'Physics - $\omega=0.85$')
axs[1,0].plot(range(epochs), lambda_100['phys'], label=r'Physics - $\omega=1.00$')
axs[1,0].set(xlabel='Epochs', ylabel='Loss', title=r'Physics-Guided Loss Term, $\mathcal{L}_G$')
axs[1,0].legend(loc='upper right'); axs[1,0].grid(True, which='both', alpha=0.25)

axs[1,1].plot(range(epochs), lambda_00['data'], label=r'Data - $\omega=0.00$')
axs[1,1].plot(range(epochs), lambda_15['data'], label=r'Data - $\omega=0.15$')
axs[1,1].plot(range(epochs), lambda_50['data'], label=r'Data - $\omega=0.50$')
axs[1,1].plot(range(epochs), lambda_85['data'], label=r'Data - $\omega=0.85$')
axs[1,1].plot(range(epochs), lambda_100['data'], label=r'Data - $\omega=1.00$')
axs[1,1].set(xlabel='Epochs', ylabel='Loss', title=r'Data Mismatch Loss Term, $\mathcal{L}_D$')
axs[1,1].legend(loc='upper right'); axs[1,1].grid(True, which='both', alpha=0.25)

axs[0,0].set(xlim=(0,epochs), ylim=(0,50))

plt.suptitle(r'$\mathcal{L}_T = \omega\mathcal{L}_G + (1-\omega)\mathcal{L}_D = \omega(\|W_d\cdot e(\theta)\|_2 + \lambda\|\theta\|_2) + (1-\omega)\| x-\hat{x}(\theta) \|_1$')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,4), sharex=True, sharey=True)

ax.plot(range(epochs), lambda_00['train'], label=r'$\omega=0.00$', c='C0')
ax.plot(range(epochs), lambda_15['train'], label=r'$\omega=0.15$', c='C1')
ax.plot(range(epochs), lambda_50['train'], label=r'$\omega=0.50$', c='C2')
ax.plot(range(epochs), lambda_85['train'], label=r'$\omega=0.85$', c='C3')
ax.plot(range(epochs), lambda_100['train'], label=r'$\omega=1.00$', c='C4')

ax.plot(range(epochs), lambda_00['valid'], ls='--', c='C0')
ax.plot(range(epochs), lambda_15['valid'], ls='--', c='C1')
ax.plot(range(epochs), lambda_50['valid'], ls='--', c='C2')
ax.plot(range(epochs), lambda_85['valid'], ls='--', c='C3')
ax.plot(range(epochs), lambda_100['valid'], ls='--', c='C4')

ax.set(xlabel='Epochs', ylabel='Loss', xlim=(0,epochs))
ax.set(title=r'$\mathcal{L}_T = \omega\mathcal{L}_G + (1-\omega)\mathcal{L}_D = \omega(\|W_d\cdot e(\theta)\|_2 + \lambda\|\theta\|_2) + (1-\omega)\| x-\hat{x}(\theta) \|_1$')
ax.grid(True, which='both', alpha=0.25)
ax.legend(loc='upper right', ncol=5)
ax.set_yscale('log')

plt.tight_layout()
plt.show()
```
